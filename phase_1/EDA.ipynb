```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing_data.csv")
X = pd.DataFrame(data)
```

```python
X = pd.get_dummies(X, columns=['property_type'] ,dtype=int)
X
```

```python
ohe_cols = [col for col in X.columns if "property_type_" in col]
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.difference(ohe_cols)

out_out = ['area_sqft', 'bedrooms', 'bathrooms', 'lot_size', 'distance_city_km', 'hospital_proximity_km', 'shopping_center_proximity_km', 'Fireplaces']
```

```python
plt.figure(figsize=(15, 10))

for i, col in enumerate(out_out):
    plt.subplot(5, 5, i+1)   
    sns.boxplot(x=X[col], color = '#5cc490')

plt.tight_layout()
plt.show()
```

```python
main_X = X.copy()
for i  in range(10):
    
    for col in out_out:
        Q1 = X[col].quantile(0.25)
        Q3 = X[col].quantile(0.75)
        IQR = Q3 - Q1
    
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
    
        X = X[
            (X[col] >= lower_bound) & 
            (X[col] <= upper_bound)
        ]
```

```python
plt.figure(figsize=(15, 10))

for i, col in enumerate(out_out):
    plt.subplot(5, 5, i+1)   
    sns.boxplot(x=X[col], color = '#5cc490')

plt.tight_layout()
plt.show()
```

```python
sns.histplot(X["SalePrice"], kde = True, color = '#5cc490')
plt.title("Sale Price Distribution")
plt.show()
```

```python
X[numeric_cols].hist(figsize =(15, 12), color = '#5cc490')
plt.show()
```

```python
plt.figure(figsize=(15, 10))

for idx, col in enumerate(numeric_cols):
    if col != "sale_price":
        plt.subplot(5, 5, idx + 1)
        plt.scatter(X[col],  X['SalePrice'], color='#5cc490', alpha=0.7)
        coeffs = np.polyfit(X[col], X['SalePrice'], 1) 
        trendline = np.poly1d(coeffs)
        plt.plot(X[col], trendline(X[col]), color='black', linewidth=2)
        plt.xlabel(col)
        plt.ylabel('sale_price')
        plt.title(f'{col} vs sale_price')

plt.tight_layout()
plt.show()
```

```python
corr = X[numeric_cols].corr()
plt.figure(figsize=(12, 10))
cmap = "coolwarm"
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap=cmap,
    vmin=-1,
    vmax=1,
)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
